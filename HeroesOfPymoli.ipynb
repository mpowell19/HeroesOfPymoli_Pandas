```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load
file_to_load = "../../RICH201901DATA3/04-Pandas/Pandas Homework/Instructions/HeroesOfPymoli/Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_df = pd.read_csv(file_to_load)
purchase_df.head()
```

## Player Count

```python
#Number of Unique Players
Total_count=len(purchase_df["SN"].unique())
Total_chart = [{"Total Players": Total_count}]
Total = pd.DataFrame(Total_chart)
Total
```

## **Purchasing Analysis (Total)**

```python
Unique=purchase_df["Item ID"].nunique()
Average=purchase_df["Price"].mean()
Count= purchase_df["Item Name"].count()
TR= Average*Count
```

```python
#Organize into a DF
Totals=[{"Number of Unique Items": Unique,
        "Average Price": Average,
        "Number of Purchases": Count,
        "Total Revenue": TR}]
Summary= pd.DataFrame(Totals)
Summary_table=Summary[["Number of Unique Items","Average Price","Number of Purchases","Total Revenue"]]
Summary_table
```

## Gender Demographics

```python
#Group by Gender
gender_groups=purchase_df.groupby("Gender")

#Total Count of Each Gender
gender_count = gender_groups['SN'].nunique()
GenderSum= pd.DataFrame(gender_count)

#Percent of Total Players
Percentage=(GenderSum["SN"]/Total_count)*100
GenderSum["Percentage"]=Percentage

#Rename
Sum_table=GenderSum.rename(columns={"SN":"Total Count","Percentage": "Percentage of Players"})

#Print DF
Sum_table
```


## Purchasing Analysis (Gender)

```python
Gender=purchase_df.groupby("Gender")["Price"].agg('sum')
Total= purchase_df.groupby("Gender").nunique()
TotalCount=Total["Purchase ID"]

#Merge the two DataFrames
Df=pd.DataFrame(Gender)
Dftwo=pd.DataFrame(TotalCount)
merge_table= pd.merge(Df, Dftwo, on="Gender")

#Add Third Column
AveragePP=(merge_table["Price"]/TotalCount)
merge_table["Average Purchase Price"]=AveragePP

#Add Last Column
AverageTotal=(merge_table["Price"]/gender_count)
merge_table["Average Total Purchase per Person"]=AverageTotal
merge_table

#Rename
New_table=merge_table.rename(columns={"Price":"Total Purchase Value","Purchase ID": "Purchase Count"})
New_table

#Rearrange
Final_table= New_table[["Purchase Count", "Average Purchase Price", "Total Purchase Value", "Average Total Purchase per Person"]]
Final_table
```

## Age Demographics

```python
#Age Bins
bins=[0,9,14,19,24,29,34,39,100]
group_names=["<10","10-14", "15-19","20-24","25-29","30-34", "35-39","40+"]
purchase_df["Age Demographics"] = pd.cut(purchase_df["Age"],bins, labels=group_names)
Updated=purchase_df

#Group by Age Bins
AgeBins= Updated.groupby("Age Demographics")["SN"].nunique()
Age=pd.DataFrame(AgeBins)

#Add column to Grouby Df
PercentofP=(Age["SN"]/Total_count)*100
Age["Percent of Players"]=PercentofP

#Rename
Age_table=Age.rename(columns={"SN":"Total Count"})
Age_table
```

## Purchasing Analysis (Age)

```python
#Group by Age
Agegroup=Updated.groupby("Age Demographics")

#Purchasing Analysis by bins
TPC= Agegroup["Price"].count()
TPV= Agegroup["Price"].sum()
APP=TPV/TPC
ATPPP= TPV/Agegroup["SN"].nunique()

final_summary=pd.DataFrame({"Purchase Count":TPC,
                    "Average Purchase Price": APP,
                    "Total Purchase Value": TPV,
                    "Avg Total Purchase Person": ATPPP})

final_summary

```

## Top Spenders

```python
TopSpenders=purchase_df.groupby("SN")

PC= TopSpenders["Purchase ID"].count()
TotalPV= TopSpenders["Price"].sum()
AveragepriceP=TotalPV/PC

Top_Spenders=pd.DataFrame({"Purchase Count":PC,
                    "Average Purchase Price": AveragepriceP,
                    "Total Purchase Value": TotalPV})
#Top_Spenders["Total Purchase Value"]=Top_Spenders["Total Purchase Value"].map("${:.2f}".format)
Top_Spenders.sort_values("Total Purchase Value", ascending=False).head()
```

## Most Popular Items

```python
#Group by Item ID/Item Name and find variables
Mostpop=purchase_df.groupby(["Item ID", "Item Name"])
PurchC= Mostpop["Purchase ID"].count()
ItemP=Mostpop["Price"].unique()
TPValue=Mostpop["Price"].sum()

#Create A DF
Most_popular=pd.DataFrame({"Purchase Count":PurchC,
                    "Item Price": ItemP,
                    "Total Purchase Value": TPValue})

Most_popular.sort_values("Purchase Count", ascending=False).head()
```
